```python
import json
from googleDriveAPI import GoogleDriveFileManager
import notebook_functions as nf

# --- Setup and Configuration ---
creds_path = 'config/google_credentials.json'
token_path = 'config/google_token.json'
config_path = 'config/config_pdf.json'
target_folder_name = "B_To Automate in Python"

try:
    drive_manager = GoogleDriveFileManager(creds_path, token_path)
    with open(config_path, 'r') as f:
        pdf_configs = json.load(f)

    # Get files from the target folder
    folder_id = drive_manager.find_folder_id_by_name(target_folder_name)
    if folder_id:
        all_files = drive_manager.list_files_in_folder(folder_id)
        files_in_folder = [(f['name'], f['id']) for f in all_files if 'pdf' in f.get('mimeType', '') or 'image' in f.get('mimeType', '')]
        if not files_in_folder:
            print(f"Warning: No PDF or image files found in '{target_folder_name}'.")
    else:
        print(f"Error: Could not find the folder '{target_folder_name}'.")
        files_in_folder = []

    # --- Create and Display the UI ---
    nf.create_interface(drive_manager, pdf_configs, files_in_folder, config_path, folder_name=target_folder_name)

except Exception as e:
    print(f"An error occurred during setup: {e}")
```
